**Dot Product**

```python
v1 = np.array([1, 3])
v2 = np.array([4, 2])

print('v1:', v1)
print('v2:', v2)
```

```python
dot = np.dot(v1, v2)
print("Dot Product v1.v2:", dot)
```

**Euclidean Distance**

```python
diff = v1 - v2
print("Difference v1 - v2:", diff)
```

```python
distance = np.linalg.norm(diff)
print("Euclidean distance between v1 and v2:", distance)
```

**Matrix Multiplication and Linear Model**

```python
X = np.array([
    [1.0, 30.0],
    [2.0, 50.0],
    [3.0, 70.0]
])

print("X:\n", X)
print("X shape:", X.shape)
```

```python
w = np.array([
    [0.5],
    [0.1]
])

print("\nw:\n", w)
print("w shape:", w.shape)
```

```python
Xw = X @ w
print("\nXw:\n", Xw)
print("Xw shape:", Xw.shape)
```

**y = Xw + b**

```python
b = 2.0
print("\nBias b:", b)
```

```python
y = Xw + b
print("\nPredictions y = Xw + b:\n", y)
print("y shape:", y.shape)
```

**Connecting this to a sample dataset**

```python
data = np.array([
    [85, 26.6, 31, 0],
    [183, 23.3, 32, 1],
    [89, 28.1, 21, 0],
    [137, 43.1, 33, 1],
])

print("Full data matrix:\n", data)
print("Data shape:", data.shape)
```

```python
X_features = data[:, :-1]
y_target = data[:, -1]

print("\nFeature matrix X:\n", X_features)
print("X shape:", X_features.shape)

print("\nTarget vector y:\n", y_target)
print("y shape:", y_target.shape)
```
